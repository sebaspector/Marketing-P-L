```python
df1['spend'] = df1['spend'].astype(float)
df1['revenue'] = df1['revenue'].astype(float)
df1['ROAS'] = df1['ROAS'].str.replace(',', '.').astype(float)
```

```python
grouped_ROAS_by_date = df1.groupby('date').agg({'revenue': 'sum', 'spend': 'sum'}).reset_index()

# Calculate the ROAS grouped by date
grouped_ROAS_by_date['ROAS'] = (grouped_ROAS_by_date['revenue'] / grouped_ROAS_by_date['spend'])
grouped_ROAS_by_date
```

```python
print(f"ROAS_mean:{grouped_ROAS_by_date['ROAS'].mean()}")
print(f"ROAS_min:{grouped_ROAS_by_date['ROAS'].min()}")
print(f"ROAS_max:{grouped_ROAS_by_date['ROAS'].max()}")
#When we look at the ratio on July 20th, itÂ´s only 1.027, compared to the 1.126 of the sample.
```

```python
grouped_ROAS_by_date_and_store = df1.groupby(['date','store']).agg({'revenue': 'sum', 'spend': 'sum'}).reset_index()

# # Calculate the ROAS grouped by date and store
grouped_ROAS_by_date_and_store['ROAS'] = (grouped_ROAS_by_date_and_store['revenue'] / grouped_ROAS_by_date_and_store['spend'])
grouped_ROAS_by_date_and_store[grouped_ROAS_by_date_and_store['date']== '2023-07-20']
```

```python
import statsmodels.api as sm

groups = df1.groupby('product')

# Iterate over each group and perform regression analysis
for group_name, group_data in groups:
  if len(group_data) >= 2:
    X = group_data['spend']  # independent variable
    y = group_data['revenue']  # dependent variable

    # Add constant term to independent variable (intercept)
    X = sm.add_constant(X)

    # Fit the linear regression model
    model = sm.OLS(y, X).fit()

    print("Group:", group_name)
    print(model.summary())
    print("\n")
  else:
      print("Skipping group:", group_name, "as it has less than 2 data points.\n")

#Linear regression model, CAMI and Leggings with higher coefficients, meaning higher ROAS, thus impacting more on the revenue.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

X = df1[['store', 'product', 'country' ,'channel']]
y = df1['ROAS']

# Add dummy variables for categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Add constant term for the intercept
X_encoded = sm.add_constant(X_encoded)

# Fit the multiple linear regression model
model = sm.OLS(y, X_encoded).fit()

# Print the model summary
print(model.summary())
```
